# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
```

# Create pytrend object

```python
pytrends = TrendReq()
```

```python
plt.rcParams["figure.figsize"] = (20,8)
```

# Define keywords, set timeframe and location

```python
# kw=['Vodafone GmbH','Deutsche Telekom AG','Telefónica Germany GmbH & Co','1&1 Drillisch AG', 'Freenet AG']
kw=['Vodafone','Deutsche Telekom','Telefonica', 'Freenet']
pytrends.build_payload(kw_list=kw,timeframe='today 5-y',geo='DE')
```

# Get interest by region

```python
result_interest_by_region = pytrends.interest_by_region()
```

```python
result_interest_by_region
```

```python
plt.rcParams["figure.figsize"] = (10,20)
result_interest_by_region.plot(kind='barh')
```

# Interest Over Time

```python
result_interest_over_time = pytrends.interest_over_time()
```

```python
result_interest_over_time
```

```python
plt.rcParams["figure.figsize"] = (20,8)
```

```python
result_interest_over_time.plot()
```

```python
result_interest_over_time['Vodafone'].plot()
```

# Get related topics

```python
words_list = ["Vodafone"]
pytrends.build_payload(words_list, cat=0, timeframe="today 5-y", geo="DE")
trend_analysis = pytrends.related_topics()
```

```python
trend_analysis.get("Vodafone").get("top")[['topic_title', 'topic_type']]
```

```python
trend_analysis.get("Vodafone").get("rising")[['topic_title', 'topic_type']]
```

# Get related queries

```python
pytrends.related_queries()
trend_analysis = pytrends.related_queries()
trend_analysis.get("Vodafone").get("top")
```

```python
trend_analysis.get("Vodafone").get("rising")
```

# Get suggestions

```python
pytrends.suggestions("Vodafone")
df = pytrends.suggestions("Vodafone")
df = pd.DataFrame(df)
df[['title','type']]
```
